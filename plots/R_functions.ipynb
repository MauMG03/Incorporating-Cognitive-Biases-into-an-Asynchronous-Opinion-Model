```python
#Plot functions 
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Plotting -x^3 ---

# Define the function
def f1(x):
    return 0.8 * x

def f3(x):
    return 0.5  * x

def f4(x):
    return 0.25 * x

def conf(x):
    delta = 0.01
    return x*(1 + delta - abs(x))/(1+delta)

# Generate x values
x3 = np.linspace(-1, 1, 200)


# Generate y values
y3_1 = f1(x3)
y3_3 = f3(x3)
y3_4 = f4(x3)
y3_conf = conf(x3)


# Create the plot
plt.figure(figsize=(10, 7))
plt.plot(x3, y3_1, color='orange', linestyle='-')
plt.plot(x3, y3_3, color='red', linestyle='-')
plt.plot(x3, y3_4, color='green', linestyle='-')
plt.plot(x3, y3_conf, color='blue', linestyle='-')

plt.xlim([-1.01, 1.01]) # Sets x-axis limits from -1 to 1
plt.ylim([-1.01, 1.01]) # Sets y-axis limits from -1 to 1
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.savefig('../images/R_region/R_functions.png', dpi=300)
plt.show()
```

```python
import math

def wave(x):
    if x >= -1 and x < -0.02:
        return (math.sin(25*x )*x + x)/2 * 0.9 - 0.01
    elif x >= -0.02 and x <= 0.02:
        return x*0.734259
    elif x <= 1 and x > 0.02:
        return (math.sin(25*x + math.pi)*x + x)/2 * 0.9 + 0.01  

# Vectorize the wave function for numpy arrays
wave_vec = np.vectorize(wave)
y_wave = wave_vec(x3)

plt.figure(figsize=(10, 7))
plt.plot(x3, y_wave, color='red', linestyle='-')
plt.xlim([-1.01, 1.01])
plt.ylim([-1.01, 1.01])
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.savefig('../images/R_region/R_wave.png')
plt.show()
```
