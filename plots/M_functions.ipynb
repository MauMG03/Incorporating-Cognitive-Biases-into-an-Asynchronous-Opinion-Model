```python
#Plot functions 
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Plotting -x^3 ---

# Define the function
def f1(x):
    return x

def f3(x):
    return 2 * x

def f4(x):
    return 5 * x

# Generate x values
x3 = np.linspace(-1, 1, 200)
# Left and right sides of the  sign function, excluding x=0
x_left = np.linspace(-1, -0.001, 200)
x_right = np.linspace(0.001, 1, 200)

# Generate y values
y3_1 = f1(x3)
y3_3 = f3(x3)
y3_4 = f4(x3)

y_left = np.sign(x_left)
y_right = np.sign(x_right)

# Create the plot
plt.figure(figsize=(10, 7))
plt.plot(x3, y3_1, color='yellow', linestyle='-')

plt.plot(x_left, y_left, color='blue')
plt.plot(x_right, y_right, color='blue')
plt.plot(0, 0, 'bo')

plt.plot(x3, y3_3, color='red', linestyle='-')

plt.plot(x3, y3_4, color='green', linestyle='-')

plt.xlim([-1.01, 1.01]) # Sets x-axis limits from -1 to 1
plt.ylim([-1.01, 1.01]) # Sets y-axis limits from -1 to 1
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.savefig('../images/M_region/M_functions.png', dpi=300)
plt.show()
```
