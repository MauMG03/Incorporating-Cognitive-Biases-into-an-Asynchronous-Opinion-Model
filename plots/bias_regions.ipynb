```python
import matplotlib.pyplot as plt
import numpy as np

# Set up the figure and axes
fig, ax = plt.subplots(figsize=(6, 6)) # Make it square

# Define the limits of the plot
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])

# --- Plotting the regions ---

# Region (I) - Yellow triangle (bottom-left, above y=0, to the left of x=0, and above y=-x)
# This region is defined by x < 0, y > -x, and y > 0
# A simple way to fill this is using `fill_between` or `fill` with polygons
# Let's define the points for the polygon (0,0), (-1,0), (-1,1)
# No, this is for the bottom left. The actual (I) is defined by:
# x in [-1, 0], y in [-1, 0] AND y > x
# It's the bottom-left quadrant but cut by the y=x line.
# Points: (-1, 0), (0, 0), (0, -1), (-1, -1) and split by y=x.
# Looking at the image, (I) is the bottom-left triangle defined by:
# x <= 0, y >= 0 and y >= -x
# Or, more simply, it's the top-left quadrant cut by y = -x
# It seems the image has (I) as the area where x < 0, y < 0 and y > x.
# So, the region (I) is below the x-axis, to the left of the y-axis, and above the diagonal y=x.
# Points: (0, 0), (-1, 0), (-1, -1), (0, -1) but bounded by y=x.
# Let's try filling polygons based on visual inspection.

# Region "M" (Top Right, Yellow)
# This is defined by x > 0, y > 0, and y < x
x_r = [0, 0, 1]
y_r = [0, 1, 1]
ax.fill_between(x_r, y_r, color='#1f78b4') # A shade of blue

# Region "M" (Top Right, Yellow)
# This is defined by x > 0, y > 0, and y < x
x_r = [0, 0, -1]
y_r = [0, -1, -1]
ax.fill_between(x_r, y_r, color='#1f78b4') # A shade of blue

# Region "R" (Top Right, Blue)
# This is defined by x > 0, y > 0, and y > x
x_m = [0, 1, 1]
y_m = [0, 1, 0]
ax.fill_between(x_m, y_m, color='#FFC300') # A shade of yellow/gold

# Region "R" (Bottom Left, Yellow)
# This is defined by x < 0, y < 0, and y > x
x_i = [0, -1, -1]
y_i = [0, -1, 0]
ax.fill_between(x_i, y_i, color='#FFC300') # Same yellow/gold

# Region "B" (Bottom Right, Pink/Red)
# This is defined by x > 0, y < 0
x_b = [0, 1, 1, 0]
y_b = [0, 0, -1, -1]
ax.fill(x_b, y_b, color='#E30B5C') # A shade of pink/red

# The remaining region in the top-left (Pink/Red)
# This is defined by x < 0, y > 0
x_tl = [-1, 0, 0, -1]
y_tl = [0, 0, 1, 1]
ax.fill(x_tl, y_tl, color='#E30B5C') # Same pink/red

# --- Add the dotted lines ---
ax.axhline(0, color='black', linestyle=':', linewidth=1)
ax.axvline(0, color='black', linestyle=':', linewidth=1)

# Add the diagonal dotted line (y = x)
#x_diag = np.linspace(-1, 1, 100)
#ax.plot(x_diag, x_diag, color='black', linestyle=':', linewidth=1)


# --- Add labels for the regions ---
# Adjust positions based on visual inspection of the original image
ax.text(-0.5, 0.05, '(I)', fontsize=14, ha='center', va='center') # I is slightly above the x-axis
ax.text(0.5, 0.75, '(M)', fontsize=14, ha='center', va='center')
ax.text(0.75, 0.25, '(R)', fontsize=14, ha='center', va='center')
ax.text(0.5, -0.5, '(B)', fontsize=14, ha='center', va='center')

# Set aspect ratio to be equal, so circles appear as circles (or squares as squares)
ax.set_aspect('equal', adjustable='box')

#plt.title('Bias Regions (Example)')
plt.grid(False) # Turn off grid lines if present

plt.savefig('../images/bias_regions.png', bbox_inches='tight', dpi=300)
plt.show()

#save plot
```
