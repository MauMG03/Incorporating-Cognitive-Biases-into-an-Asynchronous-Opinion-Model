```python
#Plot functions 
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Plotting -x^3 ---

# Define the function
def f1(x):
    return -x**3

# Define the function
def f2(x):
    return (-1/2) * x

def f3(x):
    return -3 * x

# Generate x values
x3 = np.linspace(-5, 5, 400)

# Generate y values
y3_1 = f1(x3)
y3_2 = f2(x3)
y3_3 = f3(x3)

# Create the plot
plt.figure(figsize=(10, 7))
plt.plot(x3, y3_1, color='blue', linestyle='-')
plt.plot(x3, y3_2, color='yellow', linestyle='-')
plt.plot(x3, y3_3, color='red', linestyle='-')
plt.xlim([-1, 1]) # Sets x-axis limits from -1 to 1
plt.ylim([-1, 1]) # Sets y-axis limits from -1 to 1
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.savefig('../images/B_region/B_functions.png', dpi=300)
plt.show()
```
